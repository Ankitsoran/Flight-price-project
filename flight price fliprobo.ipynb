```python
## Importing Libraries

import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver= webdriver.Chrome(r"D:/chromedriver.exe")
 #opening chrome
url='https://www.yatra.com/'
driver.get(url)
```

```python
Airline	=[] #empty list    
source=[] #empty list
Destination=[] #empty list
Dep_Time=[] #empty list
Arrival_Time=[] #empty list
Duration =[] #empty list
Total_Stops=[] #empty list
Price=[] #empty list
Date_of_Journey=[] #empty list
link=[]
```

```python
city_tag=driver.find_elements_by_xpath("//ul[@class='destlist']/li[2]")
for i in city_tag:
    i.click()
    try:
        tags=driver.find_elements_by_xpath("//a[@class='VueCarousel-slide items']")
        for i in tags:
            link.append(i.get_attribute("href"))
    except:
        pass;
```

```python

for i in link:
    driver.get(i)
    
    nm=driver.find_elements_by_xpath("//div[@class='fs-13 airline-name no-pad col-8']/span")
    for j in nm:
        Airline.append(j.text.replace("\n"," "))
    
    time=driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/div")
    for j in time:
        Dep_Time.append(j.text.replace("\n"," "))
        
    source_tag=driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")
    for j in source_tag:
        source.append(j.text.replace("\n"," "))
    
    time=driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[1]")
    for j in time:
        Arrival_Time.append(j.text.replace("\n"," "))
   
    dest_tag=driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")
    for j in dest_tag:
        Destination.append(j.text.replace("\n"," "))
   
    duration_tag=driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/p")
    for j in duration_tag:
        Duration.append(j.text.replace("\n"," "))
    
    stops_tag=driver.find_elements_by_xpath("//div[@class='stop-cont pl-13']/div")
    for j in stops_tag:
        Total_Stops.append(j.text.replace("\n"," "))
    c=0
    price_tag=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
    for j in price_tag:
        Price.append(j.text.replace("\n"," "))
        c=c+1
    
    date_tag=driver.find_element_by_id("depart-date-span")
    for j in range(c):
        Date_of_Journey.append(date_tag.text.replace("/n"," "))
```

```python
#Finding the best random state and r2_score
for i in range(100):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.20,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred_train=lr.predict(x_train)
    pred_test=lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1)==round(r2_score(y_test,pred_test)*100,1):
        print('At random state',i,',the model performs well')
        print('Training r2_score is: ',r2_score(y_train,pred_train)*100)
        print('Testing r2_score is: ',r2_score(y_test,pred_test)*100)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.20,random_state=81)
```

```python
#Initializing the instance of the model
lm= LinearRegression()
knn= KNeighborsRegressor()
dt= DecisionTreeRegressor()
rmf= RandomForestRegressor()
gb= GradientBoostingRegressor()

models=[]
models.append(('LinearRegression',lr))
models.append(('KNeighborsRegressor',knn))
models.append(('DecisionTreeRegressor',dt))
models.append(('RandomForestRegressor',rmf))
models.append(('GradientBoostingRegressor',gb))
```

```python
Model=[]
r2score=[]
mse=[]
cvs=[]
for name,model in models:
    print("****",name,"****")
    Model.append(name)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print('\n')
    AS=model.score(x_test,y_test)
    print('accuracy_score: ',AS)
    r2score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x_train,y_train,cv=5).mean()
    print('cross_val_score: ',sc)
    cvs.append(sc*100)
    print('\n')
    mse.append(mean_squared_error(y_test,y_pred))
    print('mean_squared_error ',mean_squared_error(y_test,y_pred))
    print('\n\n')
```
